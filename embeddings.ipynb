```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings = GoogleGenerativeAIEmbeddings(model='models/embedding-001')
embeddings
```

```python
text = 'This is a complete Gen AI Master Course'
result = embeddings.embed_query(text)
result
```

### 768 dimensions 
eg - 'king' -- 300-dimensional vectors[0.25, -0.13, 0.47, ..., 0.88] # 300 dimensions 
- Dimensionality : if an embedding has 300 dimensions , each vector has 300 numerical components . these dimensions collectively capture different semantic and syntactic properties of the data 

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader('solar_system.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_spiltter = RecursiveCharacterTextSplitter(chunk_size = 400, chunk_overlap=40)
final_document = text_spiltter.split_documents(docs)
final_document
```

```python
# vector embeddings and vector storeDB
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_document,embeddings)
db
```

```python
## Retrieve the results from query vectorstore db 
query = "The four inner planets—Mercury, Venus, Earth, and Mars—are terrestrial planets"
retrieved_result = db.similarity_search(query)
print(retrieved_result)
```
